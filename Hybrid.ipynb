```python
"""ender function finds the ends of the miRNA mature sequence, either from the start or stop clusters"""
def ender(reads,which_end,strand):
    libs = set(reads.values_list('lib',flat=True))
    if which_end == 'start':
        w ='start'
        seq_dict =reads.order_by('start').values('sequence','lib','start').annotate(Sum('normReads'))
    else:
        w= 'stop'
        seq_dict =reads.order_by('stop').values('sequence','lib','stop').annotate(Sum('normReads'))
    all_seq =  []
    for i in seq_dict:
        all_seq.append(i['sequence'])
    df= pd.DataFrame(index=set(all_seq),columns=libs, dtype='float64').fillna(0)
    df['sum'] = pd.Series(np.zeros(len(df.index)),index=df.index)
    df[w] = pd.Series(np.zeros(len(df.index)),index=df.index)
    for i in seq_dict:
        df.loc[i['sequence']].loc[i['lib']] += i['normReads__sum']
        df.loc[i['sequence']][w]= i[w]
        df.loc[i['sequence']]['sum'] += i['normReads__sum']
    
    df = df.sort(w)
    df_sum = df.groupby(w).sum()
    #df_sum.to_csv(path_or_buf="exceptions.csv",index_label=str(interval.mirName),mode='a')
    try:
        zero_id = df_sum['sum'].idxmax()
    except:
        zero_id = None
    
    if strand == '-':
        df_sum['diff']= pd.Series((df_sum.index.to_series()- zero_id)*-1,dtype='float64')
    else: 
        df_sum['diff']= pd.Series(df_sum.index.to_series() - zero_id,dtype='float64')
    df_gte5 = df_sum[(df_sum['diff']<= 5) & (df_sum['diff'] >=-5)].sort('diff')
    #df_gte5[w]= df_gte5[w].astype(str)
    #df_sum = df_sum.join(df_gte5['start'])
    df_gte5 = df_gte5.convert_objects(convert_numeric=True).set_index('diff')
    df_gte5= df_gte5.drop('sum',1) 
    df_final = pd.DataFrame(index= np.arange(-5,6),columns=sorted(libs), dtype='float64').fillna(0)
    for i in df_gte5.index:
        df_final.loc[i]=df_gte5.loc[i]
    return {'zero':zero_id,'summary':df_final, 'which':str(reads)}
```
